```python
%matplotlib inline

from cvm.cascade import CascadeSVM, DynamicCascadeSVM

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import svm

from sklearn.metrics import roc_auc_score
```

```python
train = pd.read_csv('data/mnist/mnist_train.csv', header=0, names=['digit'] + ['p{}'.format(i) for i in xrange(784)])
test = pd.read_csv('data/mnist/mnist_test.csv', header=0, names=['digit'] + ['p{}'.format(i) for i in xrange(784)])
```

```python
train.head()
```

```python
ytrain = 1*(train['digit'] == 5).values
Xtrain = train.drop('digit', 1).values

ytest = 1*(test['digit'] == 5).values
Xtest = test.drop('digit', 1).values
```

```python
C = 10000.0
gamma = 0.000000001

nmax = 1000

print 'training'

subsample_model = svm.SVC(C=C, gamma=gamma)
subsample_model.fit(Xtrain[:nmax], ytrain[:nmax])

print '# support vectors: {}'.format(subsample_model.n_support_)

print 'predicting'

ptrain = subsample_model.predict(Xtrain)
trainErr = np.sum(ptrain != ytrain[:nmax]) / len(ytrain[:nmax])
print 'Training error: {:.3f}'.format(trainErr)

ptest = subsample_model.predict(Xtest)
testErr = np.sum(ptest != ytest) / len(ytest)
print 'Test error: {:.3f}'.format(testErr)
```

```python
C = 1000000.0
gamma = 0.0000000001

base_model = lambda : svm.SVC(C=C, gamma=gamma)
cascade_model = CascadeSVM(base_model, nx=800, nmax=1500)

cascade_model.fit(Xtrain, ytrain)

print 'Predicting on training set...'

ptrain = cascade_model.predict(Xtrain)
trainErr = np.sum(ptrain != ytrain) / len(ytrain)
print 'Training error: {:.3f}'.format(trainErr)

print 'Predicting on test set...'

ptest = cascade_model.predict(Xtest)
testErr = np.sum(ptest != ytest) / len(ytest)
print 'Test error: {:.3f}'.format(testErr)
```
